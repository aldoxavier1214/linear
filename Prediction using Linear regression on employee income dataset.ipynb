```python
import numpy as np
# Function to detect outliers using Z-score
def detect_outliers_zscore(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    return z_scores > threshold

# Select only numeric columns for outlier detection
numeric_columns = emp.select_dtypes(include=[np.number]).columns

# Detect and mark outliers in the DataFrame
outliers = emp[numeric_columns].apply(detect_outliers_zscore)

# Drop rows with outliers (you may choose to handle outliers differently based on your specific use case)
df_no_outliers = emp[~outliers.any(axis=1)]
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select numeric columns for normalization (excluding the target variable if it exists)
numeric_columns = emp.drop("Total_Reimbursement", axis=1).select_dtypes(include=[np.number]).columns

# Initialize the Min-Max Scaler
scaler = MinMaxScaler()

# Fit and transform the data using the scaler
emp[numeric_columns] = scaler.fit_transform(emp[numeric_columns])

# Display the normalized DataFrame
print(emp.head())
```

```python
# Convert 'Year' column to datetime format
emp['Year'] = pd.to_datetime(emp['Year'])

# Extract date-time features
emp['Month'] = emp['Year'].dt.month
emp['Day'] = emp['Year'].dt.day
emp['DayOfWeek'] = emp['Year'].dt.dayofweek
emp['Hour'] = emp['Year'].dt.hour
emp['Minute'] = emp['Year'].dt.minute

# Drop the original 'Year' column (optional)
emp.drop('Year', axis=1, inplace=True)
```

```python
# One-hot encode categorical features (assuming 'Type_of_yr' is a categorical variable)
df = pd.get_dummies(emp, columns=['Type_of_yr'], drop_first=True)
df
```

```python
# Create a new feature 'Total_Income_to_Benefits_Ratio'
df['Total_Income_to_Benefits_Ratio'] = df['Total_Income'] / df['Total_Benefits']
```

```python
df['Total_Income_to_Benefits_Ratio']
```

```python
# Binning 'Income' into three categories: Low, Medium, and High
income_bins = [0, 50000, 100000, np.inf]
income_labels = ['Low', 'Medium', 'High']
df['Income_Category'] = pd.cut(df['Income'], bins=income_bins, labels=income_labels)
```

```python
# Apply log transformation to 'Income' column
df['Log_Income'] = np.log1p(df['Income'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# Assuming 'emp' is the DataFrame after feature engineering
# Drop irrelevant columns and set the target variable (Total_Reimbursement)
X = emp.drop(['Total_Reimbursement'], axis=1)
y = emp['Total_Reimbursement']

# One-hot encode 'Type_of_yr_Financial' column
encoder = OneHotEncoder(sparse=False, drop='first')  # Use drop='first' to avoid multicollinearity
X_encoded = pd.DataFrame(encoder.fit_transform(X[['Type_of_yr']]))
X_encoded.columns = encoder.get_feature_names_out(['Type_of_yr'])
X = pd.concat([X.drop(['Type_of_yr'], axis=1), X_encoded], axis=1)

# Perform data imputation to handle missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Predict 'Total_Reimbursement' on the testing data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2) Score:", r2)
```
